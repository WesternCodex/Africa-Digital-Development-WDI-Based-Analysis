```python
df_ti = pd.read_csv('Africa_Digital_Development.csv')
print("Number of Records:", len(df_ti))
print("Total Region :", df_ti["Region"].nunique())
print("Total Country :", df_ti["Country Name"].nunique())
print("Total Country Code :", df_ti["Country Code"].nunique())
print("Year Range: ", df_ti["Years"].min(), "-", df_ti["Years"].max())

population = df_ti[
    df_ti["Indicator Name"] == "Population, total"
].copy()

# Remove missing and zero values
population = population[
    (population["value"].notna()) &
    (population["value"] > 0)
]

# Find the latest year that actually contains data
latest_year = population["Years"].max()

total_population = (
    population[population["Years"] == latest_year]["value"]
    .sum()
)

print("Latest Valid Year:", latest_year)
print(f"Total Population: {total_population:,.0f}")

```

```python
def fun_kpi(ind_name):
    """ avoid repeating code """
    indicator = df_ti[
        (df_ti["Indicator Name"] == ind_name)
        & (df_ti["value"] != 0)
    ]

    return round(indicator["value"].mean(), 2)



print(f"Average Internet Usage: {fun_kpi('Individuals using the Internet (% of population)')}%")
print(f"Average GDP Per Capital: ${fun_kpi('GDP per capita (current US$)')}")
print(f"Average GDP Growth: {fun_kpi('GDP growth (annual %)')}%")
print(f"Average Mobile Subscription: {fun_kpi('Mobile cellular subscriptions (per 100 people)')}")
print(f"No of indicators {df_ti['Indicator Name'].nunique()}")



```

```python
# Making visuals
internet = df_ti[df_ti["Indicator Name"] == "Individuals using the Internet (% of population)"].copy()
mobile = df_ti[df_ti["Indicator Name"] == "Mobile cellular subscriptions (per 100 people)"].copy()
gdp = df_ti[df_ti["Indicator Name"] == "GDP per capita (current US$)"].copy()
gdp_growth = df_ti[df_ti["Indicator Name"] == "GDP growth (annual %)"].copy()
population = df_ti[df_ti["Indicator Name"] == "Population, total"].copy()


def fun_indicator(name_): 
    name_trend = (
        name_.groupby("Years")["value"]
        .mean()
        .reset_index()
    )

    return name_trend

internet_trend = fun_indicator(internet)
mobile_trend = fun_indicator(mobile)
gdp_trend = fun_indicator(gdp)
gdpGrowth_trend = fun_indicator(gdp_growth)
population_trend = fun_indicator(population)
```

```python
# check internet and gdp correlation
internet_latest = internet[internet["Years"] == latest_year][
    ["Country Name", "value"]
].rename(columns={"value": "Internet"})

gdp_latest = gdp[gdp["Years"] == latest_year][
    ["Country Name", "value"]
].rename(columns={"value": "GDP"})

internet_latest = internet_latest[internet_latest["Internet"] != 0]  # Treating zeros as missing values and remove them from the correllation
gdp_latestgdp_latest = gdp_latest[gdp_latest["GDP"] != 0]
merged = internet_latest.merge(
    gdp_latest,
    on="Country Name"
)

corr = merged["Internet"].corr(merged["GDP"])

print("Correlation:", round(corr, 2))
internet_latest.head()
# Close to 1 → strong positive relationship.
# Around 0 → little or no relationship.
# Negative → inverse relationship.
```

```python
# check internet and mobile correlation

mobile_latest = mobile[mobile["Years"] == latest_year][
    ["Country Name", "value"]
].rename(columns={"value": "Mobile"})
mobile_latest = mobile_latest[mobile_latest["Mobile"] != 0]

merged = internet_latest.merge(
    mobile_latest,
    on="Country Name"
)

corr = merged["Mobile"].corr(merged["Internet"])

print("Correlation:", round(corr, 2))
internet_latest.head()
# Close to 1 → strong positive relationship.
# Around 0 → little or no relationship.
# Negative → inverse relationship.
```
